```python
books_dir = '/Users/valentin/Downloads/Pack.Thermomix.Balisé.30.Livres.40.Magazines'
```

```python
def title_filter(title):
    return title[-2] if len(title) > 1 and len(title[-2]) > 7 else title[-1]
```

```python
list_books = os.listdir(books_dir)[2:]
list_books = [title_filter(x.split(' - ')) for x in list_books]
list_books
```

```python
import Levenshtein
from operator import getitem
```

```python
for book in unique_titles:
    print "===== {} =====".format(book)
    score, title = min([(Levenshtein.distance(book, book_), book_) for book_ in list_books], key=lambda x:getitem(x, 0))
    if score < 10:
        print title, score
```
